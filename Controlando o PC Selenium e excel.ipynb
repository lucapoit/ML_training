```python
#Programa pra atualizar minha planilha de gastos com o valor atual do dolar e do euro

import xlsxwriter
import openpyxl
import os 
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pyautogui
from selenium.webdriver.common.by import By
from datetime import datetime
from string import ascii_uppercase as alc


#pegando a data atual para saber em qual coluna adicionar o valor do dolar/euro
a = datetime.today().strftime('%Y-%m-%d')
a = a.split('-')
a = int(a[1])
meses = ('Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro')
mes_atual = meses[a-1]
print(mes_atual)

#ação para pegar o valor do dolar e euro hoje
Navegador = webdriver.Chrome()
Navegador.get('https://www.google.com/')

pyautogui.sleep(2)

#cuidado com o metodo find_element que as vezes é no plural ou singular
Navegador.find_element(By.NAME, 'q').send_keys('dolar hoje') 
pyautogui.sleep(2)
Navegador.find_element(By.NAME, 'q').send_keys(Keys.RETURN) 
pyautogui.sleep(2)
valor_dolar= Navegador.find_elements(By.XPATH, '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]')[0].text
pyautogui.sleep(2)
valor_dolar = valor_dolar.split(',')
valor_dolar = '.'.join(valor_dolar)
valor_dolar = float(valor_dolar)

Navegador.find_element(By.NAME, 'q').send_keys('') 
pyautogui.sleep(1)
pyautogui.press('tab')
pyautogui.sleep(1)
pyautogui.press('enter')
pyautogui.sleep(1)

Navegador.find_element(By.NAME, 'q').send_keys('euro hoje') 
pyautogui.sleep(1)
Navegador.find_element(By.NAME, 'q').send_keys(Keys.RETURN) 
pyautogui.sleep(2)

valor_euro = Navegador.find_elements(By.XPATH, '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]')[0].text
valor_euro = valor_euro.split(',')
valor_euro = '.'.join(valor_euro)
valor_euro = float(valor_euro)


#importante colocar a barra dupla pois a primeira é funcional, como no \n para quebra de linha por exemplo
caminho = 'C:\\Users\\luca_\\Desktop\\2024\\Teste.xlsx'
wb = openpyxl.load_workbook(caminho)
sh = wb.active
att_coluna = 0
dol_linha= 0
eur_linha = 0

#achar a posição que preciso inserir os novos valores do dolar e do euro
for col in sh.iter_cols(min_row=3, max_col=10, max_row=3, values_only=True):    
    if col[0] == mes_atual:
        break
    att_coluna+=1
    
att_coluna = alc[att_coluna]

for row in sh.iter_rows(min_row=0, max_col=0, max_row=40, values_only=True):
    dol_linha+=1    
    if row[0] == 'USD/BRL': 
        break
        
for row in sh.iter_rows(min_row=0, max_col=0, max_row=40, values_only=True):
    eur_linha+=1    
    if row[0] == 'EUR/BRL': 
        break
        
dol_cell = ''.join([str(att_coluna),str(dol_linha)])
eur_cell = ''.join([str(att_coluna),str(eur_linha)])

print(dol_cell)
print(eur_cell)
    
sh[dol_cell].value = valor_dolar
sh[eur_cell].value = valor_euro

wb.save(caminho)

os.startfile(caminho)
```
